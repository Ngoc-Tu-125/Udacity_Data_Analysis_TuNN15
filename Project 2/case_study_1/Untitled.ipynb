```python
import pandas as pd

red_df = pd.read_csv('winequality-red.csv', sep=';')
white_df = pd.read_csv('winequality-white.csv', sep=';')

print(red_df.shape)

print(white_df.shape)

white_df.duplicated().sum()
```

```python
import numpy as np

#create color columns
color_red = np.repeat('red', red_df.shape[0])
```

```python
red_df['color'] = color_red
red_df.head()
```

```python
color_white = np.repeat('white', white_df.shape[0])
white_df['color'] = color_white
white_df.head()
```

```python
wine_df = pd.concat([red_df, white_df])

wine_df.head()

wine_df.to_csv('wine_df_combined.csv')
```

```python
bin_edges = [2.72, 3.11, 3.21, 3.32, 4.01]
bin_names = ['high', 'mod_high', 'medium', 'low']

w_df['acidity_levels'] = pd.cut(w_df['pH'], bin_edges, bin_names)

w_df.head()
```

```python
w_df.groupby('acidity_levels').mean().quality
```

```python
low_alcohol = w_df.query('alcohol < 10.3')

high_alcohol = w_df.query('alcohol >= 10.3')

num_samples = w_df.shape[0]

num_samples == low_alcohol['quality'].count() + high_alcohol['quality'].count()
```
