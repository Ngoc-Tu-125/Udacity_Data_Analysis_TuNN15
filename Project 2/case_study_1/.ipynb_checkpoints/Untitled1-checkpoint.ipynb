```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

wine_df = pd.read_csv('wine_df_combined.csv')

wine_df.head()
```

```python
colors = ['red', 'green']

wine_df.groupby('color')['quality'].mean().plot(kind='bar', title='Average Wine Quality by Color', color=colors, alpha = .6)

plt.xlabel('Colors', fontsize=18)
plt.ylabel('Quality', fontsize=18)
```

```python
counts = wine_df.groupby(['quality', 'color']).count()['pH']
counts
```

```python
counts.plot(kind='bar', title='Counts by Wine Color and Quality', color=colors, alpha = .7)
```
